```python
import sys
sys.path.append('../src')

import scipy.io as sio
import numpy as np
import matplotlib.pyplot as plt
from config import FS, N_CHANNELS

# Load data
data = sio.loadmat('../data/S17_E1_A1.mat')
emg = data['emg']
labels = data['restimulus']

print("EMG shape:", emg.shape)
print("Labels shape:", labels.shape)
print("Unique gestures:", np.unique(labels))

samplesToPlot = 60 * FS

plt.figure(figsize=(12, 4))
plt.plot(emg[:samplesToPlot, 0])
plt.title("Raw EMG - Channel 1 (first 5 seconds)")
plt.xlabel("Sample")
plt.ylabel("Amplitude")
plt.tight_layout()
plt.show()
```
