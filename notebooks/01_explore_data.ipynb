```python
import scipy.io as sio
import numpy as np
import matplotlib.pyplot as plt


# Load the file
data = sio.loadmat('S17_E1_A1.mat')

#Looking what's inside
print(data.keys())

#Looking what the emg data is
emg = data['emg']
labels = data['restimulus']

print("EMG shape: ", emg.shape)
print("Labels shape: ", labels.shape)
print("Unique gestures: ", np.unique(labels))

# Sampling rate for Ninapro DB2 is 2000 Hz
fs = 2000

# Plot a minute of channel 1
samplesToPlot = 60 * fs

plt.figure(figsize=(12, 4))
plt.plot(emg[:samplesToPlot, 0])
plt.title("Raw EMG - Channel 1 (The first minute)")
plt.xlabel("Sample")
plt.ylabel("Amplitude")
plt.tight_layout()
plt.show()

fig, axes = plt.subplots(12, 1, figsize=(12, 16), sharex=True)

for i in range(12):
    axes[i].plot(emg[:samplesToPlot, i], linewidth=0.5)
    axes[i].set_ylabel(f'Ch {i+1}', fontsize=8)
    axes[i].set_ylim(-8e-5, 8e-5)

axes[-1].set_xlabel('Sample')
fig.suptitle('All 12 EMG Channels - First minute', fontsize=12)
plt.tight_layout()
plt.show()
```
